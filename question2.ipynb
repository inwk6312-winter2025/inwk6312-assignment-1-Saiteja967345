````python
from tabulate import tabulate
# Import more libraries if needed
import json
from tabulate import tabulate

# Load the dataset
with open("hfxbus.json", "r") as file:
    data = json.load(file)

# Extract relevant columns from the dataset
table_data = []
for item in data:
    table_data.append([
        item.get("agency_id", "N/A"),  # Extract 'Agency Id'
        item.get("route_long_name", "N/A"),  # Extract 'Route Long Name'
        item.get("route_type", "N/A"),  # Extract 'Route Type'
        item.get("route_color", "N/A")  # Extract 'Route Color'
    ])

# Define table headers
headers = ["Agency Id", "Route Long Name", "Route Type", "Route Color"]

# Print the table in GitHub style
print(tabulate(table_data, headers=headers, tablefmt="github"))






````

# 2D (30 points)

- How many of the traffics are of **http** service and have **source bytes** greater than **199** ?

````python
# Answer is 271

import json

# Load the dataset from the JSON file
with open('intrusion.json', 'r') as file:
    data = json.load(file)

# Initialize a counter for HTTP traffic with source bytes > 199
http_count = 0

# Iterate through the dataset and count HTTP traffic with source bytes > 199
for record in data:
    # Check if the service is HTTP and source bytes are greater than 199
    if record.get('service') == 'http' and record.get('src_bytes', 0) > 199:
        http_count += 1

# Print the count of HTTP traffic with source bytes > 199
print(http_count)

# If you want to check for a specific output, you can do:
if http_count == 860:  # Replace 860 with the expected count if needed
    print("HTTP traffic count with source bytes > 199 is 860.")
else:
    print("HTTP traffic count with source bytes > 199 is not 860, it is:", http_count)


````

# 2E -- **BONUS** (50 points)

- Using Matplotlib library, plot traffic data are of **http service** ,  have **source bytes greater than 150** and **source bytes less than 199** on a graph

To make your graph look indentical, use Keyword arguments:

``` linestyle='dashed', marker='o', markerfacecolor='red', markersize=10 ```

````python
import matplotlib.pyplot as plt
import json

# Load the dataset from the JSON file
with open('intrusion.json', 'r') as file:
    data = json.load(file)

# Initialize lists to hold the source bytes for plotting
source_bytes = []

# Iterate through the dataset and collect source bytes for HTTP traffic
for record in data:
    # Check if the service is HTTP and source bytes are within the specified range
    if record.get('service') == 'http' and 150 < record.get('src_bytes', 0) < 199:
        source_bytes.append(record['src_bytes'])

# Create a range for the x-axis (1 to number of records)
x_values = range(1, len(source_bytes) + 1)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x_values, source_bytes, linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)

# Add titles and labels
plt.title('HTTP Traffic with Source Bytes between 150 and 199')
plt.xlabel('Record Index')
plt.ylabel('Source Bytes')
plt.grid(True)

# Show the plot
plt.show()




````

# 3A - 40 Points
Open the XML file and print all the **authors** in the book collection, as well as the **name** of the book and the **price**.

````python
import xmltodict

import xml.etree.ElementTree as ET

# Function to parse the XML and print book details
def extract_book_details(xml_file):
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Iterate through each book and print details
    for book in root.findall('book'):
        title = book.find('title').text
        price = book.find('price').text
        authors = [author.text for author in book.findall('author')]  # Assuming multiple authors can exist

        # Print the book details
        print(f"Book Title: {title}")
        print(f"Price: {price}")
        print(f"Authors: {', '.join(authors)}")
        print('-' * 40)

# Example usage
xml_file = 'books.xml'  # Replace with the path to your XML file
extract_book_details(xml_file)



````

# 3B - 30 Points
Tabulate the output in Task 3A. Your table should look like this:
```
| Author               | Title                                  |   Price |
|----------------------|----------------------------------------|---------|

| Ralls, Kim           | Midnight Rain                          |    5.95 |
| Corets, Eva          | Maeve Ascendant                        |    5.95 |
```

````python
import xml.etree.ElementTree as ET
from tabulate import tabulate

# Load the XML file
tree = ET.parse("books.xml")  # Replace with your XML filename
root = tree.getroot()

# Extract book details
table_data = []
for book in root.findall("book"):  # Adjust the tag name if necessary
    author = book.find("author").text if book.find("author") is not None else "Unknown"
    title = book.find("title").text if book.find("title") is not None else "No Title"
    price = book.find("price").text if book.find("price") is not None else "N/A"

    table_data.append([author, title, price])

# Define headers
headers = ["Author", "Title", "Price"]

# Print the table in GitHub style
print(tabulate(table_data, headers=headers, tablefmt="github"))




````

How many **unique** authors are they in the XML file ?

````python
# Answer is 9
import xml.etree.ElementTree as ET

# Load the XML file
tree = ET.parse("books.xml")  # Replace with your XML filename
root = tree.getroot()

# Extract all unique authors
unique_authors = set()
for book in root.findall("book"):  # Adjust the tag name if necessary
    author = book.find("author").text if book.find("author") is not None else "Unknown"
    unique_authors.add(author)  # Add author to the set

# Print the number of unique authors
print(f"Number of unique authors: {len(unique_authors)}")




````
